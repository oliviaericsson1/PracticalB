```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = "testdata.csv"
df = pd.read_csv(data)
```

```python
sns.scatterplot(data=df, x="query_time", y="Total", hue="Vector_DB", s=80)
plt.title("Query Time vs. Total Score by Vector DB")
plt.xlabel("Query Time (seconds)")
plt.ylabel("Total Score")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x="embedding_model", y="Total", estimator='mean', ci=None)
plt.title("Average Total Score by Embedding Model")
plt.ylabel("Avg Total Score (out of 15)")
plt.xlabel("Embedding Model")
plt.ylim(0, 15)
plt.tight_layout()
plt.show()
```
